```python
# Imports
import numpy as np
import matplotlib.pyplot as plt

# Create a new figure of size 8x6 points, using 100 dots per inch
plt.figure(figsize=(8,6), dpi=100)
plt.title("cos and sin")

# Create a new subplot from a grid of 1x1
plt.subplot(111)

X = np.linspace(-np.pi, np.pi, 256, endpoint=True)
C = np.cos(X)
S = np.sin(X)

plt.xlabel("x")
plt.ylabel("y")

plt.xlim(-np.pi,np.pi)
plt.ylim(-1,1)

plt.yticks([-1,-0.5, 0, 0.5, 1 ])

plt.xticks([-np.pi,-np.pi/2,0, np.pi/2, np.pi],[r'$ -\pi $ ',r' $ -\frac{ \pi }{ 2 } $ ' ,'$0$' ,r' $ \frac{ \pi }{ 2 } $ ' ,r' $ \pi $ ' ])

#Plot cosine
plt.plot(X, C, linewidth=2, linestyle='dashdot', color='blue', label="y = cos(x)")
plt.plot(X, S, linewidth=1, linestyle='dashed', color='red', label="y = sin(x)")

plt.annotate("Annotation", xy=(0,0),  xytext=(0.4,0), arrowprops=dict(arrowstyle='->',lw=1.5))
plt.legend(loc='best')

# Show result on screen
plt.show()
```

# Étape 6 : Interagir avec Matplolib (1/2)
Matplotlib fournit un certain nombre d'interacteurs de base (bouton, slidder, etc.) qui s'appuit sur l'environment disponible sur votre plateforme (e.g. Qt, MAC OS, etc.). voir : https://matplotlib.org/3.2.1/users/interactive.html

Ci dessous, un exemple d'une simple <strong>QMainWindow</strong> qui intègre un </strong>FigureCanvasQTAgg</strong>, un Widget qui hérite de QWidget (et donc ses propriétés, signals, slots et méthodes) et qui contient un <strong>matplotlib.figure</strong> pour afficher des plots.



```python
import sys
import time

import numpy as np

from PyQt5.QtWidgets import *
from PyQt5.QtGui import *
from PyQt5.QtPrintSupport import *
from PyQt5.QtCore import *

from matplotlib.backends.backend_qt5agg import FigureCanvasQTAgg as FigureCanvas
from matplotlib.backends.backend_qt5agg import NavigationToolbar2QT as NavigationToolbar
import matplotlib.pyplot as plt

class Window(QMainWindow):
    def __init__(self):
        super().__init__()
        self.count = 0
        self.color_text = "blue"
        self.container = QWidget()
        layout = QVBoxLayout( self.container )
        
        # A Qwidget embedding a matplotlib figure
        self.canvas  = FigureCanvas( plt.figure( figsize=(5, 3)) )
        layout.addWidget( self.canvas )
        
        # add the matplotlib toolbar
        self.addToolBar( NavigationToolbar( self.canvas, self ) )
        
        self.button = QPushButton( "red" )
        self.button.clicked.connect( self.update_canvas )
        layout.addWidget( self.button )

        self.ax = self.canvas.figure.subplots()
        
        self.update_canvas()
        
        self.setCentralWidget( self.container )

    def update_canvas(self):
        self.count += 1
        
        self.ax.clear()
        t = np.linspace( 0, 10, 101 )
        self.ax.plot( t, np.sin(t + self.count ), c= self.color_text )
        self.ax.set_xlabel( "Time (s)" )
        self.ax.set_ylabel( "Temperature (celcius)" )
        self.canvas.draw()
        self.canvas.repaint()  # required for MACOS
        
        self.color_text = "blue" if self.color_text == "red" else "red"
        self.button.setText( self.color_text )    


if __name__ == "__main__":
    qapp = QApplication(sys.argv)
    win = Window()
    win.show()
    qapp.exec_()
```

# Étape 6 : Interagir avec Matplotlib (2/2)
Il est possible de récupérer les déplacements de la souris dans un plot. Ci dessous, un programme qui vous vous permet de manipuler une courbe de Bézier à l'aide de quatre points de control.
<ul>
    <li> Familiarisez vous avec le code, en particulier les méthodes utilisées pour récupérer les évenements souris </li>
    <li> Modifier <strong>on_press</strong>, <strong>on_move</strong>, <strong>on_release</strong> pour manipuler interactiver les quatre points de control. On utilisera pour cela, la méthode <strong>get_control_point_index_from </strong> (à implémenter) pour mettre à jour <strong>self.index</strong> (id du point de control à manipuler) et <strong>self.control</strong> un array qui contient les coordonnées x,y des quatre points de control. Par exemple self.control[2,0], self.control[2,1] sont les coordonnées x, y du troisième point de control. </li>
    
</ul>



```python
import sys

import numpy as np

from PyQt5.QtWidgets import *
from PyQt5.QtGui import *
from PyQt5.QtPrintSupport import *
from PyQt5.QtCore import *

from matplotlib.backends.backend_qt5agg import FigureCanvasQTAgg as FigureCanvas
from matplotlib.backends.backend_qt5agg import NavigationToolbar2QT as NavigationToolbar
import matplotlib.pyplot as plt

class Window(QMainWindow):
    
    #################
    def __init__(self):
        super().__init__()
        
        # A Qwidget embedding a matplotlib figure
        self.canvas  = FigureCanvas( plt.figure( figsize=(5, 3)) )
        self.ax = self.canvas.figure.subplots()
        
        # add the matplotlib toolbar
        self.addToolBar( NavigationToolbar( self.canvas, self ) )
        
        self.index = -1 # index of the current control point
        
        # initial coordinates of the control points
        self.control = np.array( [ [1.0,1.0],[1.0,9.0],[9.0,1.0],[9.0,9.0] ] ) 
        
        
        self.update_canvas()
        
        # get mouse press, mouse move (drag + move), mouse release
        self.canvas.mpl_connect("button_press_event"  , self.on_press)
        self.canvas.mpl_connect("motion_notify_event" , self.on_move)
        self.canvas.mpl_connect("button_release_event", self.on_release)
        
        self.setCentralWidget( self.canvas )
    
    
    ####################
    # return a bezier curve with four control points
    def cubic_bezier_curve( self, t , control ):
        t_array = np.array( [t,t])
        res = np.zeros((len(t),2))
        for i in [0,1]: 
            res[:,i] = (1 - t)**3 * control[0,i] + 3 * t*(1-t)**2 * control[1,i] + 3 * (1-t)*t**2 * control[2,i] + t**3 * control[3,i]
        return  res 
    
    
    ####################
    # inputs: coordinates (x,y) in the figure (not in the QWidget)
    # ouput: index of the control points res \in {0,1,2,3} or -1
    # Given the position p = (x,y) of the mouse event in the frame of reference
    # of the matplotlib figure, this method returns the closest control point ci
    # if distance(p, ci) < dist_max and -1 otherwise.
    ###################
    def get_control_point_index_from(self, x, y ):
        res = -1        # no control points found
        max_dist = 0.5
        l=[]
        for k in self.control:
            l.append(np.linalg.norm(k -(x,y)))
        arg=np.argmin(l)
        
        if l[arg]<max_dist:
            res=arg
        
        return res
            
    ####################
    def update_canvas(self):
        # update bezier points
        t = np.linspace(0,1, 30)
        self.bezier = self.cubic_bezier_curve( t, self.control )
        
        # update the plot
        self.ax.clear()
        self.ax.plot(self.bezier[:,0], self.bezier[:,1])
        self.ax.scatter(self.control[:,0], self.control[:,1], c= "orange" )
        self.ax.set_xlim(0,10), self.ax.set_ylim(0,10)
        self.ax.set_xticks([]), self.ax.set_yticks([])
        self.ax.set_title ('Cubic Bézier Curve')
        
        
        # repaint
        self.canvas.draw()
        self.canvas.repaint()  # required for MACOS
    
      
    ####################
    def on_press(self, event):
        self.index = self.get_control_point_index_from(event.xdata, event.ydata)
        print("press -- xdata: ", event.xdata, " ydata: ", event.ydata, " x: ", event.x, " y: ", event.y )
        print(self.index)
    
    ####################
    def on_move(self, event):
        if self.index != -1:
            self.control[self.index] = [event.xdata, event.ydata]
            self.update_canvas()
    
    ####################
    def on_release(self, event):
        print("on_release")
        self.index = -1
        print(self.control)
    

##########################
#          MAIN          #
##########################
if __name__ == "__main__":
    qapp = QApplication(sys.argv)
    win = Window()
    win.show()
    qapp.exec_()
```
